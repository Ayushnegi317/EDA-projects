```python
import pandas as pd
import plotly.express as pl
```

```python
df=pd.read_csv("/content/Credit card transactions - India - Simple.csv")
```

```python
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
df
```

```python
df1=df.sort_values("Date",ascending=True)
df2=df1.drop("index",axis=1)
```

```python
df.isnull().sum()
```

```python
df1.info()
```

```python
df3=df2.groupby(['Gender','Exp Type']).size().reset_index(name='Count')
df3
```

```python
pl.histogram(df3,x="Exp Type",y='Count',color='Gender',barmode="group")
```

**Observation:**
Females has highest transaction in food Section Compared to mens 

```python
df4=df2.groupby(['Date','City','Amount','Gender']).size().reset_index(name="count")
df4
```

```python
pl.histogram(df4,x="Date",y="Amount",color='Gender',barmode='group',histfunc="max")
```

```python
df5=df2.groupby(['City','Amount']).size().reset_index(name="count")
df5
```

```python
df6=df5.nlargest(20,['Amount'])
```

```python
pl.histogram(df6,x="City",y="Amount")
```

**Observation:**
Greater Mumbai has highest Transaction in top cities of transactions and delhi,ahmedabad,bengaluru has most transaction from enitre cities in data
