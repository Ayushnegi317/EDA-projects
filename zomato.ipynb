```python
#now finding the highest transaction in No. of Countries
currency=df.groupby(["Country","Currency"]).size().reset_index(name="No. of Transaction")
```

```python
pl.pie(currency,values='No. of Transaction',labels="Currency",color="Country",title="Most Transaction in Country")
```

**Observation**

Most Transaction have been done in India and then USA

```python
#highest transaction in Cities
currency1=df.groupby(["City","Currency"]).size().reset_index(name="No. of Transaction")
```

```python
pl.histogram(currency1,x="City",y="No. of Transaction",title="Most Transaction in City")
```

**Observation**
Most Transaction in have been done in cities are 

New Delhi

Noida

Gurgaon
